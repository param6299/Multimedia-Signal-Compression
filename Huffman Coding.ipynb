```python
# Reading the text
text = open('Lab 2.txt','r')
x = list(text.read().lower())
y = list(set(x))
n = len(x)
```

```python
# Finding the probability
count = 0
p = []
freq = []
for i in range(len(y)):
    for r in range(len(x)):
        if y[i] == x[r]:
            count +=1
        else:
            continue
    p.append(count/n)
    freq.append(count)
    count = 0
print(p)
```

```python
# Forming a dataframe
import pandas as pd
df = pd.DataFrame({'Character':y , 'Frequency':freq, 'Probability':p})
df
```

```python
# Descending order
for i in range(len(p)):
    for j in range(i+1):
        if p[j]<p[i]:
            t = p[j]
            p[j] = p[i]
            p[i] = t

p
```

```python
# Ascending order
for i in range(len(p)):
    for j in range(i+1):
        if p[j]>p[i]:
            t = p[j]
            p[j] = p[i]
            p[i] = t
            
p
```
